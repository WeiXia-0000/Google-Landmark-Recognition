# Map img ID to their paths

```python
# prompt: Given a folder, I want to recursively loop through the folder and look for all files ending in ".jpg", and create a mapping pair (file_name, file_path), where file_name is the name without extension and file_path is the path of the file. Save this to a csv with two columns, id and path

import os

def create_file_mapping(folder):
  """
  Recursively loops through a folder and creates a mapping pair (file_name, file_path) for each .jpg file.

  Args:
    folder: The path to the folder to search.

  Returns:
    A list of tuples, where each tuple contains the file name without extension and its path.
  """

  file_mapping = []

  for root, dirs, files in os.walk(folder):
    for file in files:
      if file.endswith(".jpg"):
        file_name = os.path.splitext(file)[0]
        file_path = os.path.join(root, file)
        file_mapping.append((file_name, file_path))
        # print(file_mapping)
        # return

  return file_mapping

# Example usage:
# folder = "/content/drive/MyDrive/cs444-final-project/project/data/selected_train"
folder = "/Users/ruipenghan/Desktop/Academics/11. SP 2024/CS 444/project/data/small_data/train"
file_mapping = create_file_mapping(folder)

# Save the file mapping to a CSV file:
import pandas as pd

df = pd.DataFrame(file_mapping, columns=["id", "path"])
save_to = "/Users/ruipenghan/Desktop/Academics/11. SP 2024/CS 444/project/train_img_id_to_path_mapping_drive.csv"
df.to_csv(save_to, index=False)
```

# Convert Data to PT

```python
import os
import torch
from torchvision import transforms
from PIL import Image
from tqdm import tqdm


def convert_images_to_tensor(image_dir, save_dir, transform):
    # List all JPEG images in the directory
    image_files = [f for f in os.listdir(image_dir) if f.endswith('.jpg')]

    for image_file in tqdm(image_files, desc="Converting images"):
        image_path = os.path.join(image_dir, image_file)
        image = Image.open(image_path).convert('RGB')
        tensor = transform(image)
        # Save the tensor; replaces '.jpg' with '.pt'
        tensor_filename = image_file.replace('.jpg', '.pt')
        torch.save(tensor, os.path.join(save_dir, tensor_filename))


# Directory containing your JPEG images
train_image_dir = '/content/drive/MyDrive/cs444-final-project/project/data/small_data/train'
test_image_dir = '/content/drive/MyDrive/cs444-final-project/project/data/small_data/test'

# Directory where you will save the .pt files
train_save_dir = '/content/drive/MyDrive/cs444-final-project/project/data/small_data/train_pt'
test_save_dir = '/content/drive/MyDrive/cs444-final-project/project/data/small_data/test_pt'
os.makedirs(test_save_dir, exist_ok=True)


test_transform = transforms.Compose([
    transforms.Resize((256, 256)),
    transforms.CenterCrop(224),
    transforms.ToTensor()
])


# Run the conversion
convert_images_to_tensor(test_image_dir, test_save_dir, test_transform)
```

```python
train_transform = transforms.Compose([
    transforms.Resize((256, 256)),
    # transforms.CenterCrop(224),
    transforms.RandomCrop(224),
    transforms.ToTensor()
])

convert_images_to_tensor(train_image_dir, train_save_dir, train_transform)
```
